# MaStar Spectra

For this assignment, you are asked to identiy the spectral type for a set of stars.  The data for assignment is from the MaNGA Stellar Library (MaStar), part of SDSS IV.  This data set is described on this [webpage](https://www.sdss.org/dr15/mastar/) and in a paper [here](https://iopscience.iop.org/article/10.3847/1538-4357/ab3ebc/pdf).  Feel free to explore!

```python
# get working directory
import os
assignment_dir = os.getcwd()

# load spectra
import astropy.io.fits as fits
goodspec = fits.open(assignment_dir+ '/mastar-goodspec.fits')[1].data

# load selected object IDs
with open('mystery_ids.txt') as f:
    mystery_ids = f.readlines()
mystery_ids = [x.strip() for x in mystery_ids]

print(mystery_ids)
```

## Plotting Spectra

Below, I have provided some basic Python to plot the spectra of the selected objects.  You may need to manipulate and/or replot each of the spectra using different parameters in order to make an accurate identification. 

```python
# plot one of the spectra
ID = mystery_ids[0]
# there may be multiple spectra for each object
visits = (goodspec["MANGAID"] == ID)
wl = goodspec[visits]["WAVE"][0]
flux = goodspec[visits]["FLUX"][0]

fig = plt.figure()
plt.title(ID)
plt.plot(wl, flux, '-', color='black')
plt.ylabel("flux (1e-17 erg/s/cm$^2$/Angstrom)")
plt.xlabel("wavelength (Angstroms)")
plt.show()
```

```python
# plot all of the spectra
for ID in mystery_ids:
    visits = (goodspec["MANGAID"] == ID)
    wl = goodspec[visits]["WAVE"][0]
    flux = goodspec[visits]["FLUX"][0]

    fig = plt.figure()
    plt.title(ID)
    plt.plot(wl, flux, '-', color='black')
    plt.ylabel("flux (1e-17 erg/s/cm$^2$/Angstrom)")
    plt.xlabel("wavelength (Angstroms)")
    plt.show()
```
